```python
#import packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf
pdf = matplotlib.backends.backend_pdf.PdfPages("visual3Nov.pdf")
```

```python
excel_file='Input Files/Flight Schedule 3 Nov.xlsx'
Nov3=pd.read_excel(excel_file)
```

```python

Nov3.columns
```

```python
routes={}
for i in range(len(Nov3["TAIL"])):
    routes[Nov3["TAIL"][i]]=[]
print(routes.keys())
print(len(routes))
```

```python
for i in range(len(Nov3["TAIL"])):   
    routes[Nov3["TAIL"][i]].append([Nov3["DEP"][i], Nov3["RTDIST"][i], Nov3["ARR"][i], Nov3["RTAIST"][i]])
```

```python
for c in routes.keys():
    routes[c].sort(key=lambda x: x[1])
print(routes['VTAIF'])
```

```python

for c in routes.keys():
    fig = plt.figure(figsize=(12,4))
    ax = fig.add_subplot(111)
    A = range(2*len(routes[c]))
    B = [6,4]*int(len(A)/2)
    plt.scatter(A,B)
    for i,j in zip(A,B):
        if i%2==0:
            ax.annotate('%s,%s' %(routes[c][int(i/2)][0],routes[c][int(i/2)][1].split()[1]), xy=(i,j))
            ax.arrow(i,6, 1, -2,   width=0.015, head_width=0.1, head_length=0.25)
        else:
            ax.annotate('%s,%s' %(routes[c][int(i/2)][2],routes[c][int(i/2)][3].split()[1]), xy=(i,j))
    
    
    plt.suptitle("Aircraft: "+str(c))
    plt. xticks(A, " ")
    plt. yticks(range(4,7), ["Arrival", " ", "Departure"])
    pdf.savefig( fig )
    plt.show()
    
    
pdf.close()
```
